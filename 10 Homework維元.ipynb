### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```




````python
 #1 df.loc[ '2013-01-01', 'A']: 只會選取到一個值，這個值的row是'2013-01-01'，而column則是'A'
 #2 df.loc[ '2013-01-01', ['A', 'B'] ]: 會選取到row是'2013-01-01'，而column是'A','B'的值
 #3 df.loc[ '2013-01-01':'2013-01-02', 'A' ]:會選取到row從'2013-01-01'到'2013-01-02'，而column是'A'的值
 #4 df.loc[ '2013-01-01':'2013-01-05', 'A':'C']: 會選取到row從'2013-01-01'到'2013-01-05'，而column從'A'到'C'的值
````

### 2. 請根據提供的資料，選擇出下列的要求：

```
# - select the first 3 rows.
# - select the odd rows. (index = 1, 3, 5)
# - select the last 2 columns.
# - select the even columns. (index = 0, 2, 4)
```


````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
# - select the first 3 rows.
print(df.loc[0:2,:])
# - select the odd rows. (index = 1, 3, 5).
print(df.iloc[lambda x: x.index % 2 == 1])
# - select the last 2 columns.
print(df.iloc[:,[-2,-1]])
# - select the even columns. (index = 0, 2, 4)
print(df.iloc[:,[0,2]])
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
# - 1. filtered by first column > 20?
df.loc[df.loc[:,0]>20,[0]]
````

````python
# - 2. filtered by first column + second column > 50
df['0+1']=df.loc[:,0]+df.loc[:,1]
df.loc[df['0+1']>50,['0+1']]
````

````python
# - 3. filtered by first column < 30 or second column > 30
df.loc[(df.loc[:,0]<30)|(df.loc[:,1]>30),[0,1]]
````

````python
# - 4. filtered by total sum of row > 100
df['sum']=df.loc[:,0]+df.loc[:,1]+df.loc[:,2]+df.loc[:,3]
df.loc[df['sum']>100,['sum']]
````
